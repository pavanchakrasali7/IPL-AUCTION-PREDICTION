### Import Pycaret Library, that automatically fits teh data to different models and compares them based on many loss functions and accuracy

```python
from pycaret.regression import *
```

### Pass the data and target feature to the pycaret model and specify required features for tuning

```python
reg = setup(data = ipl_full, 
            target = 'Value',
            normalize = True,
            session_id=123)
```

### comparing the results of accuracy and loss functions of different models

```python
compare_models()
```

### Create a variable to save the parameters considered for teh catboost regressor, since it has a relatively high accuracy and low RMSe value

```python
catboost = create_model('catboost') 
```

### Fetching the parameters of the catboost regressor

```python
catboost.get_params()
```

### Hyper parameter tuning for the catboost regressor model

```python
catboost_tune=tune_model(catboost)   
```

```python
print(catboost_tune)
```

### Evaluating the cat boost model

```python
evaluate_model(catboost_tune)
```

### Use the model to predict the IPL Player value and check the error values

```python
predict_model(catboost_tune);
```

### Finalise the catboost model

```python
final_catboost = finalize_model(catboost_tune)
```
